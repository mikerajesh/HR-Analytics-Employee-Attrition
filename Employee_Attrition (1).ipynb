```python
import pandas as pd

url = "https://raw.githubusercontent.com/mikerajesh/HR-Analytics-Employee-Attrition/main/WA_Fn-UseC_-HR-Employee-Attrition.csv"

df = pd.read_csv(url)

df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
constant_cols = [
    col for col in df.columns
    if df[col].nunique() == 1
]

print(constant_cols)
```

```python
df_clean = df.drop(
    columns=constant_cols
)
```

```python
df_clean.to_csv(
    "cleaned_attrition.csv",
    index=False
)
```

```python
df_clean = pd.read_csv(
    "cleaned_attrition.csv"
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = df_clean.select_dtypes(
    include='object'
).columns

for col in cat_cols:
    df_clean[col] = le.fit_transform(
        df_clean[col]
    )
```

```python
corr_target = df_clean.corr()['Attrition'].abs()

corr_target.sort_values(
    ascending=False
)
```

```python
selected_features = corr_target[
    corr_target > 0.05
].index

print(selected_features)
```

```python
df_selected = df_clean[
    selected_features
]

df_selected.head()
```

```python
df_selected.to_csv(
    "selected_features.csv",
    index=False
)
```
