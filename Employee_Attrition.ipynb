```python
import pandas as pd

url = "https://raw.githubusercontent.com/mikerajesh/HR-Analytics-Employee-Attrition/main/WA_Fn-UseC_-HR-Employee-Attrition.csv"

df = pd.read_csv(url)

df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
constant_cols = [
    col for col in df.columns
    if df[col].nunique() == 1
]

print(constant_cols)
```

```python
df_clean = df.drop(
    columns=constant_cols
)
```

```python
df_clean.to_csv(
    "cleaned_attrition.csv",
    index=False
)
```
